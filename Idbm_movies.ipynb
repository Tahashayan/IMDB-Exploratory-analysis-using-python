```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Idbm_movies.xlsx')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop(['Year_Release', 'Genre'], axis = 1)
df.head(5)
```

```python
df = df.rename(columns={'Title':'Movie_Name'})
df.head(5)
```

```python
idbm_top_movie = df[['Movie_Name', 'Rating', 'Votes']][df["Rating"] == df["Rating"].max()]
idbm_top_movie.head()
```

```python
df_top10 = df.head(20)
plt.figure(figsize=(20,25))
sns.scatterplot(x="Movie_Name", y="Votes", data=df_top10)
plt.xlabel("Average Rating")
plt.ylabel("Number of Votes")
plt.xticks(rotation=50)
plt.title("Average Rating vs Number of Votes for Movies")
plt.show()
```

```python
plt.figure(figsize=(11,10))
sns.barplot(x="Rating", y="Votes", data=df)
plt.xlabel("Rating")
plt.ylabel("Number of Votes")
plt.title("Top Movies by Number of Votes")
plt.xticks(rotation=0)
plt.show()
```

```python
df_top20 = df.head(20)
plt.figure(figsize=(17,22))
sns.histplot(x="Votes", y="Rating", data=df_top20, kde=True)
plt.xlabel("Votes")
plt.ylabel("Rating")
plt.title("Histogram of Rating for Movies")
plt.show()
```
